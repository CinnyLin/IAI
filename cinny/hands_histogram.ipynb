# Hands Histogram
Create histograms by player to understand player strategy

### import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### import data

```python
df = pd.read_csv('total_hands_summary.csv', index_col=0)
df.head(1)
```

### explore

```python
df_Eddie = df[df['player']=='Eddie']
df_Eddie.head(1)
```

```python
plt.hist(df_Eddie['total_hands'], label='Eddie', alpha=0.8)
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.hist(df_Eddie['total_hands'], label='Eddie', alpha=0.8)
plt.legend()
```

```python
# sorted by total games played (as in doc)
players = ['Pluribus', 'ORen', 'MrWhite', 'MrPink', 'MrOrange', 'MrBrown', 'MrBlue',
          'MrBlonde', 'Joe', 'Gogo', 'Hattori', 'Eddie', 'Budd', 'Bill']
```

### plot total_hands

```python
for player in players:
    df_player = df[df['player']==player]
    plt.hist(df_player['total_hands'], label=player, alpha=0.8)
    plt.legend()
    text = f'{player}_total_hands'
    plt.show()
    plt.savefig(f'hands_histogram/total_hands/{text}.png')
```

```python
n = np.arange(1,15)
fig, ax = plt.subplots(4, 4, figsize=(20,10))

for i in n:
    player = players[i-1]
    df_player_total_hands = df[df['player']==player]['total_hands']
    if i<5:#k=1-4
        ax[0,i-1].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[0,i-1].legend()
    elif i<9: #k=5-8
        ax[1,i-5].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[1,i-5].legend()
    elif i<13: #k=9-12
        ax[2,i-9].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[2,i-9].legend()
    else: #k=13-16
        ax[3,i-13].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[3,i-13].legend()
    plt.savefig(f'hands_histogram/summaries/total_hands_summary.png')
```

```python
n = np.arange(1,15)
fig, ax = plt.subplots(4, 4, figsize=(20,10))

for i in n:
    player = players[i-1]
    df_player_total_hands = df[df['player']==player]['total_hands']
    if i<5:#k=1-4
        ax[0,i-1].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[0,i-1].legend()
        ax[0,i-1].set_xlim(0, 300)
        ax[0,i-1].set_ylim(0, 20)
    elif i<9: #k=5-8
        ax[1,i-5].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[1,i-5].legend()
        ax[1,i-5].set_xlim(0, 300)
        ax[1,i-5].set_ylim(0, 20)
    elif i<13: #k=9-12
        ax[2,i-9].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[2,i-9].legend()
        ax[2,i-9].set_xlim(0, 300)
        ax[2,i-9].set_ylim(0, 20)
    else: #k=13-16
        ax[3,i-13].hist(df_player_total_hands, label=player, alpha=0.8)
        ax[3,i-13].legend()
        ax[3,i-13].set_xlim(0, 300)
        ax[3,i-13].set_ylim(0, 20)
    plt.savefig(f'hands_histogram/summaries/total_hands_summary_scaled.png')
```

### plot hands_won_percentage

```python
for player in players:
    df_player = df[df['player']==player]
    percent_won = df_player['total_hands_won']/df_player['total_hands']
    plt.hist(percent_won, label=player, alpha=0.8)
    plt.legend()
    text = f'{player}_hands_won_percentage'
    plt.show()
    plt.savefig(f'hands_histogram/hands_won_percentage/{text}.png')
```

```python
n = np.arange(1,15)
fig, ax = plt.subplots(4, 4, figsize=(20,10))

for i in n:
    player = players[i-1]
    df_player = df[df['player']==player]
    df_player_percent_won = df_player['total_hands_won']/df_player['total_hands']
    if i<5:#k=1-4
        ax[0,i-1].hist(df_player_percent_won, label=player, alpha=0.8)
        ax[0,i-1].legend()
    elif i<9: #k=5-8
        ax[1,i-5].hist(df_player_percent_won, label=player, alpha=0.8)
        ax[1,i-5].legend()
    elif i<13: #k=9-12
        ax[2,i-9].hist(df_player_percent_won, label=player, alpha=0.8)
        ax[2,i-9].legend()
    else: #k=13-16
        ax[3,i-13].hist(df_player_percent_won, label=player, alpha=0.8)
        ax[3,i-13].legend()
    plt.savefig(f'hands_histogram/summaries/hands_won_percentage_summary.png')
```
