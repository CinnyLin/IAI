```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df.csv')
df.head()
```

## explore

```python
df_40 = df[df['game']==40]
players = list(set(df_40['player']))

plt.figure(figsize=(30,20))
for player in players:
    df_40_player = df_40[df_40['player']==player]
    hands = list(df_40_player['hand'])
    money = list(df_40_player['start_stack'])
    plt.plot(hands, money, label=player)
plt.legend(fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
games = list(set(df['game']))
plt.figure(figsize=(30,20))
for game in games:
    df_game = df[df['game']==game]
    players = list(set(df_game['player']))
    for player in players:
        df_game_player = df_game[df_game['player']==player]
        hands = list(df_game_player['hand'])
        money = list(df_game_player['start_stack'])
        plt.plot(hands, money, label=player)
    plt.legend(fontsize=20)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    text = f'game_{game}_summary'
    plt.title(text, fontsize=20)
    plt.savefig(f'game_plot_summary/{text}.png')
    plt.clf()
```
